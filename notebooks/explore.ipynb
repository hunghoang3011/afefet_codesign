```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
print("\n" + "="*60)
print("Figure 5h: SNN training accuracy")
print("="*60)

df_5h = pd.read_excel(xls, sheet_name='Figure5h')
print(f"Shape: {df_5h.shape}")
print(f"Columns: {df_5h.columns.tolist()}")
print("\nFirst 10 rows:")
print(df_5h.head(10))
```

```python
import json
import os

# ========== EXTRACT DEVICE PARAMETERS ==========

# 1. Switching time từ Figure 1c
df_1c = pd.read_excel(xls, sheet_name='Figure1c')

# Find characteristic switching time (từ paper: τ_AFE=170ns, τ_FE=398ns)
# Ở normalized polarization = 0.5 (midpoint)
afe_data = df_1c['Hf0.13Zr0.87O2']
fe_data = df_1c['Hf0.5Zr0.5O2']

# Tìm index gần 0.5 nhất
afe_idx = (afe_data - 0.5).abs().idxmin()
fe_idx = (fe_data - 0.5).abs().idxmin()

tau_afe = 10 ** df_1c.loc[afe_idx, 'logt']  # convert từ log
tau_fe = 10 ** df_1c.loc[fe_idx, 'logt']

print(f"AFE switching time: {tau_afe*1e9:.1f} ns")
print(f"FE switching time: {tau_fe*1e9:.1f} ns")

# 2. LTP/LTD parameters từ Figure 4a
df_4a = pd.read_excel(xls, sheet_name='Figure4a')

# Lấy conductance values (bỏ column 'Cycle')
conductance_data = df_4a.drop('Cycle', axis=1).values

# Calculate cycle-to-cycle variation
variations = []
for i in range(len(conductance_data)-1):
    diff = np.abs(conductance_data[i] - conductance_data[i+1])
    var = diff / (conductance_data[i] + 1e-10)
    variations.append(np.mean(var))

cycle_variation = np.mean(variations)
print(f"\nCycle-to-cycle variation: {cycle_variation*100:.2f}%")

# 3. SNN baseline accuracy từ Figure 5h
df_5h = pd.read_excel(xls, sheet_name='Figure5h')
final_accuracy = df_5h['acc'].iloc[-1]
best_accuracy = df_5h['acc'].max()

print(f"\nSNN Final accuracy: {final_accuracy*100:.2f}%")
print(f"SNN Best accuracy: {best_accuracy*100:.2f}%")

# ========== SAVE PARAMETERS ==========
device_params = {
    'switching_time': {
        'afe_ns': float(tau_afe * 1e9),
        'fe_ns': float(tau_fe * 1e9)
    },
    'voltage_levels': [3.5, 4.0, 4.5, 5.0, 5.5],  # từ paper Fig 5d
    'variation': float(cycle_variation),
    'baseline_accuracy': {
        'final': float(final_accuracy),
        'best': float(best_accuracy)
    },
    'energy': {
        'baseline_pj': 0.1,  # từ paper
        'pulse_width_us': 1.0,
        'voltage_v': 5.0
    }
}

# Save to JSON
os.makedirs('../data/processed', exist_ok=True)
with open('../data/processed/device_params.json', 'w') as f:
    json.dump(device_params, f, indent=2)

print("\n✓ Parameters saved to data/processed/device_params.json")
print("\nSummary:")
print(json.dumps(device_params, indent=2))
```

```python
# ========== PLOT KEY FIGURES ==========
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1: Switching dynamics
ax1 = axes[0]
ax1.plot(10**df_1c['logt']*1e6, df_1c['Hf0.5Zr0.5O2'], label='FE (Hf0.5Zr0.5O2)')
ax1.plot(10**df_1c['logt']*1e6, df_1c['Hf0.13Zr0.87O2'], label='AFE (Hf0.13Zr0.87O2)')
ax1.set_xlabel('Pulse Width (μs)')
ax1.set_ylabel('Normalized Polarization')
ax1.set_xscale('log')
ax1.legend()
ax1.set_title('Switching Dynamics (Fig 1c)')
ax1.grid(True, alpha=0.3)

# Plot 2: LTP behavior (first 100 cycles)
ax2 = axes[1]
# Average conductance across devices
mean_conductance = conductance_data.mean(axis=1)
ax2.plot(df_4a['Cycle'], mean_conductance*1e6, 'o-', markersize=3)
ax2.set_xlabel('Pulse Number')
ax2.set_ylabel('Conductance (μS)')
ax2.set_title('LTP/LTD Behavior (Fig 4a)')
ax2.grid(True, alpha=0.3)

# Plot 3: SNN training curve
ax3 = axes[2]
ax3.plot(df_5h['epoch'], df_5h['acc']*100, 'o-', markersize=4)
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Accuracy (%)')
ax3.set_title('SNN Training (Fig 5h)')
ax3.axhline(y=97.8, color='r', linestyle='--', alpha=0.5, label='Target: 97.8%')
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../results/data_overview.png', dpi=150, bbox_inches='tight')
print("\n✓ Figure saved to results/data_overview.png")
plt.show()
```
